```python
# This import will allow us to create the scatter plot needed, each one has a specific role 
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# below I will import the data from the file
insurance_data = pd.read_csv("insurance.csv")
```

```python
# On the below block we will extract the columns of 'age' and 'charges' with all their rows
x = insurance_data.iloc[:,:1].values
y = insurance_data.iloc[:,-1].values
```

```python
# Now here we will plot the chart using the methods from matplotlib
plt.scatter(x, y, color = 'r')
plt.xlabel("Age (years)")
plt.ylabel("Charges (£)")
plt.title("How age affects insurance costs")
```

```python
# Below code has been written by using linear_model.LinearRegression() for a fit model.
age_affects_insurance = LinearRegression()
age_affects_insurance.fit(x,y)
```

```python
# Plotting the data using the fit model
y_predict = age_affects_insurance.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_predict, color = 'r')
plt.xlabel("Age (years)")
plt.ylabel("Charges (£)")
plt.title("How age affects insurance costs")
```

It appears that there is a positive relationship between the independent variable "Age" and the dependent variable "Charges". As the age increases, the dependent variable also increases.

```python
# Below there is code for a best fit model using seaborn library
plt.figure()
sns.scatterplot(data= insurance_data, x='age', y= 'charges')
plt.plot(x, y_predict)
plt.xlabel("Age (years)")
plt.ylabel("Charges (£)")
plt.title("How age affects insurance costs")
plt.show()
```
